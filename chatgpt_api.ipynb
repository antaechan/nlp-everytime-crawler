```python
import openai
import os
from dotenv import load_dotenv
load_dotenv()
openai.api_key = os.getenv("openai_key")
```

```python
import pandas as pd
dataset_all = pd.read_csv("ret_data/total.csv").dropna(axis=0)
dataset_all
```

```python
dataset_all['context'] = dataset_all['context'].apply(lambda x: ' '.join(x.split('\n')))
dataset_all['context'].apply(lambda x: ('\n') in x).sum()
```

```python
messages = []
query = "아래의 문장들을 읽고 여성/가족과 관련된 혐오표현이면 1, 남성과 관련된 혐오 표현이면 2, 성소수자는 3, 인종/국적은 4, 연령은 5, 지역은 6, 종교는 7, 기타혐오는 8, 악플/욕설은 9, 혐오표현이 없는 글은 10, 개인을 지칭하는 혐오 표현은 11을 출력해줘, 리스트 형식으로 출력해줘"
messages.append({"role": "system", "content": query})
for i, context in enumerate(dataset_all['context'].to_list()):
    messages.append({"role": "user", "content": context})
    if i >= 30:
        break
    
response = openai.ChatCompletion.create(model="gpt-3.5-turbo", messages=messages)
chatgptLabel = response["choices"][0]["message"]["content"]
print(chatgptLabel)
# dataset_all.loc[i, "chatgptLabel"] = chatgptLabel
```
